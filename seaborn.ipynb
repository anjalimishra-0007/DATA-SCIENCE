```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r'C:\Users\dell\Desktop\Hi-tech\student_data_modified.xlsx')
```

```python
sns.lineplot(data=df, x='study_hours', y='exam_score', hue='placement_status')
plt.show()
```

```python
sns.barplot(data=df, x='sleep_hours', y='exam_score',hue='placement_status', estimator= np.mean)
plt.show()
```

```python
sns.histplot(data=df, x='attendance', hue='placement_status', bins=25)
plt.show()
```

```python
sns.scatterplot(data=df, x='previous_score', y='exam_score', hue='placement_status')
plt.title('Penguin Bill Dimensions')
plt.show()
```

```python
sns.boxplot(data = df, x='placement_status', y='previous_score')
plt.show()
```

```python
df_heat = df[['attendance', 'assignments_completed', 'sleep_hours', 'study_hours']].corr()
sns.heatmap(df_heat, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
