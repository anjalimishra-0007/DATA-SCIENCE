```python
lst =[1,2,3,'hello','c',[2,4]]
print(lst)

# append()
lst.append(3)
print(lst)

# extend()
lst.extend([2,3,4])
print(lst)

# insert()
lst.insert(2,"value")
print(lst)
```

```python
lst=[2,5,7,8,11,34]
lst.append(5)
print(lst)
lst.extend(["shreya","anjali"])
print(lst)
lst.insert(0,"a")
print(lst)

# pop()
lst.pop()
print(lst)

# remove()
lst.remove(2)
print(lst)
# index()
# print(lst.index())
# count()
# print(lst.count(3))

# sort()
lst=[3,9,6,8,1,4]
lst.sort()
print(lst)


# sum,min,max
```
