```python
class person:
  def __init__(self,h,w,age):
    self.height=h
    self.weight=w
    self.age=age

class Rahul(person):
  def __init__(self,g,name,b,h,w,age):
    super().__init__(h,w,age)

    self.gender=g
    self.name=name
    self.bank_balance=b
obj=Rahul("male","Rahul",100,5,55,18)
print(obj.age)
```

```python
class complex:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,other):
    # return complex(self.x+other.x,self.y+other.y)
    a=self.x+other.x
    b=self.y+other.y
    result=complex(a,b)
    return result

  def display(self):
    print(self.x,'+i',self.y)

a=complex(2,3)
b=complex(4,5)

c=a+b
# print(c.x,c.y)
c.display()
```
